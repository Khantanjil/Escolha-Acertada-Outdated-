```python
from bs4 import BeautifulSoup
import pandas
import requests

# 21 pages
l = []
a = []
b = []
for page in range(0,22):
    base_url = "http://www.escolhaacertada.pt/pt-PT/Imoveis/-/-/-/-/-/-/-/-/-/-/-/-/0/0/20/1/"
    if page == 0:
        continue
    #rint(base_url + str(page))
    r = requests.get(base_url+str(page), headers={
        'User-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'
    })
    c = r.content
    soup = BeautifulSoup(c, "html.parser")
    all = soup.find_all("div", {"class":"row"})
    #print(len(all))
    for item in all:
        d = {}
        price = soup.find_all("div", {"class":"price"})
        name = soup.find_all("a", {"class":"trackevent"})
    for prices in price:
        prices = prices.text
        prices = " ".join(prices.split())
        a.append(prices)
        d["Preco"] = a
    """for names in name:
        names = names.text
        names = " ".join(names.split())
        b.append(names)
        d["Ímovel"] = b
        print(b)
    """
    l.append(d)
```
