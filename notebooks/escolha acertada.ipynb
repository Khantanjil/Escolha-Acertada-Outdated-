```python
from bs4 import BeautifulSoup
import pandas
import requests

# 21 pages
l = []
for page in range(0,22):
    base_url = "http://www.escolhaacertada.pt/pt-PT/Imoveis/-/-/-/-/-/-/-/-/-/-/-/-/0/0/20/1/"
    if page == 0:
        continue
    #rint(base_url + str(page))
    r = requests.get(base_url+str(page), headers={
        'User-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'
    })
    c = r.content
    soup = BeautifulSoup(c, "html.parser")
    all = soup.find_all("div", {"class":"row"})
    #print(len(all))
    for item in all:
        d = {}
        for i in item.find_all("div", {"class":"price"}):
            sell = i.find_all("i", {"class":"fa fa-home"})
            #print(sell)
            sell = i.text
            #print(sell)
            sell = " ".join(sell.split())
            #print(sell)
            d["Preço"] = sell
            l.append(d)
        for i in item.find_all("div", {"class":"info"}):
            nome = i.find_all("a", {"class":"trackevent"})
            nome = nome[0].text
            #print(nome) 
            d["Imóvel"] = nome
            l.append(d)
        for i in item.find_all("div", {"class":"info"}):
            beds = i.find_next('ul').find_all('li')
            beds = beds[0]
            #print(beds)
            for i in beds.find_all("li"):
                #print(i)
                a = i.find_all("i", {"class":"icon-bedrooms"})
                #print(a)
        l.append(d)
        #print(d)
        df = pandas.DataFrame(l)
        escolha_acertada = df.dropna()
        escolha_acertada = escolha_acertada.drop_duplicates()
```
